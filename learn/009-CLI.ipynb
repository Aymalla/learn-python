# Package Installation

```python
# Install Packages

# pip install package_name==version
!pip install colorama

# Uninstall Package
!pip uninstall -y colorama

# List installed packages
!pip freeze
!pip freeze > requirements.txt

# Install from requirements file
!pip install -r requirements.txt

# Uninstall all packages from requirements file
!pipe uninstall -y -r requirements.txt

# Uninstall all packages
!pip freeze | xargs pip uninstall -y
```

# Python Virtual Environment

Python Virtual Environment is a tool to create isolated Python environments. It is a way to create a separate environment for each project. It solves the “Project X depends on version 1.x but, Project Y needs 4.x” dilemma and keeps your global site-packages directory clean and manageable.

```python
# Create a virtual environment
!python -m venv myenv

# Activate the virtual environment
!source myenv/bin/activate

# Deactivate the virtual environment
!deactivate

# List installed packages
!pip freeze
!pip freeze > requirements.txt

# Install from requirements file
!pip install -r requirements.txt

# remove the virtual environment
!rm -rf myenv
```
